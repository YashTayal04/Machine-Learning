# Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Reading Testing Data

```python
dfx=pd.read_csv("./Training Data/Train.csv")
```

```python
data=dfx.values
x=data[:][:,:5]
y=data[:][:,5]
print(x.shape)
print(y.shape)
```

```python
plt.scatter(x[:,0],y)
plt.title("Feature 1")
plt.show()
```

```python
plt.scatter(x[:,1],y)
plt.title("Feature ")
plt.show()
```

```python
plt.scatter(x[:,2],y)
plt.title("Feature 3")
plt.show()
```

```python
plt.scatter(x[:,3],y)
plt.title("Feature 4")
plt.show()
```

```python
plt.scatter(x[:,4],y)
plt.title("Feature 5")
plt.show()
```

# Training Model

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
```

```python
model.fit(x,y)
```

# Prediction

```python
bias = model.intercept_
coeff = model.coef_

print(bias)
print(coeff)
```

```python
test_data=pd.read_csv("./Testing Data/Test.csv")
output=model.predict(test_data)
test_data=test_data.values
```

```python
plt.scatter(x[:,0],y,label='data')
plt.plot(test_data[:,0],output,color='orange',label='prediction')
plt.legend()
plt.title("Feature 1")
plt.show()
plt.scatter(x[:,1],y,label='data')
plt.plot(test_data[:,1],output,color='orange',label='prediction')
plt.legend()
plt.title("Feature 2")
plt.show()
plt.scatter(x[:,2],y,label='data')
plt.plot(test_data[:,2],output,color='orange',label='prediction')
plt.legend()
plt.title("Feature 3")
plt.show()
plt.scatter(x[:,3],y,label='data')
plt.plot(test_data[:,3],output,color='orange',label='prediction')
plt.legend()
plt.title("Feature 4")
plt.show()
plt.scatter(x[:,4],y,label='data')
plt.plot(test_data[:,4],output,color='orange',label='prediction')
plt.legend()
plt.title("Feature 5")
plt.show()
```

```python
model.score(x,y)
```
